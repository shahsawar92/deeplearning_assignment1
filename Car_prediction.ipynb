```python
mean=train_data.iloc[:,0:14].mean(0)
train_data.iloc[:,0:14]-=mean
std=train_data.iloc[:,0:14].std(0)
train_data.iloc[:,0:14]/=std
train_data.iloc[:,0:14]-=mean
test_data.iloc[:,0:14]/=std
```

```python
mean_label=train_label.mean(0)
train_label-=mean_label
std_label=train_label.std(0)
train_label/=std_label
test_label-=mean_label
test_label/=std_label
```

```python
test_label
```

```python
train=np.array(train_data.iloc[:]).astype('float32')
test=np.array(test_data.iloc[:]).astype('float32')
train_l=np.array(train_label.astype('float32'))
test_l=np.array(test_label.astype('float32'))
```

```python
test
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python

    model=models.Sequential()
    model.add(layers.Dense(128,activation='relu',input_shape=(train.shape[1],)))
    model.add(layers.Dense(62,activation='relu'))
    model.add(layers.Dense(32,activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
```

```python
h=model.fit(train,train_l,epochs=10,batch_size=100)
```

```python
test_l,test_a=model.evaluate(test,test_l)
```
