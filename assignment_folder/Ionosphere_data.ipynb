```python
import pandas as pd
import numpy as np
from tensorflow.keras import models
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
```

```python
data=pd.read_csv('./ionosphere_data.csv')
```

```python
data
```

```python
(data.isnull()).any()
```

```python
label=data['label']
```

```python
data=data.drop('label',axis=1)
```

```python
new_label=pd.get_dummies(label)
```

```python
from sklearn import preprocessing

x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data = pd.DataFrame(x_scaled)
data
```

```python
np.random.seed(1122)
mask=np.random.rand(len(data))<0.6
train_data=data[mask]
test_data=data[~mask]
train_l=new_label[mask]
test_l=new_label[~mask]
```

```python
x_train=np.array(train_data.iloc[:]).astype('float32')
x_test=np.array(test_data.iloc[:]).astype('float32')
y_train=np.array(train_l.astype('float32'))
y_test=np.array(test_l.astype('float32'))
```

```python
model= models.Sequential()
```

```python
    model.add(layers.Dense(8,activation='relu',input_shape=(data.shape[1],)))
    model.add(layers.Dense(8,activation='relu'))
    model.add(layers.Dense(6,activation='relu'))
    model.add(layers.Dense(2,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```
