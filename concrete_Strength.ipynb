```python
from sklearn import preprocessing

x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data = pd.DataFrame(x_scaled)
data
```

```python
train_label=data['Concrete compressive strength(MPa, megapascals) ']
```

```python
mask=np.random.rand(len(data))<0.5
train_data=data[mask]
test_data=data[~mask]
validate_data= test_data.sample(frac = 0.4)
test =test_data.drop(validate_data.index)
train_l=train_label[mask]
testa=train_label[~mask]
validate_label= testa.sample(frac = 0.4)
test_label=test_data.drop(validate_data.index)
```

```python
print(len(test))
print(len(validate_data))
print(len(train_data))
train_data.shape[1]
```

```python
print(len(test_label))
print(len(validate_label))
print(len(train_l))
```

```python
x=np.array(train_data.iloc[:]).astype('float32')
x_test=np.array(test.iloc[:]).astype('float32')
x_validate=np.array(validate_data.iloc[:]).astype('float32')
y_validate=np.array(validate_label.astype('float32'))
y_train=np.array(train_l.astype('float32'))
y_test=np.array(test_label.astype('float32'))
```

```python
x_test
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model=models.Sequential()
```

```python
    model.add(layers.Dense(8,activation='relu',input_shape=(8,)))
    model.add(layers.Dense(8,activation='relu'))
    model.add(layers.Dense(10,activation='relu'))
    model.add(layers.Dense(6,activation='relu'))
    model.add(layers.Dense(1))
```

```python
    model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
```

```python
model.fit(x,y_train,epochs=100,batch_size=10)
```

```python
model.evaluate(x_validate,y_validate)
```

```python
predictions = model.predict(x_test)
predictions
```
